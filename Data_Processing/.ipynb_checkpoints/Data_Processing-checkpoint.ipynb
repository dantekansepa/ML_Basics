# Data Exploration Techniques Part A and Part B

*1. Dimensionality Check;*
The shape attribute returns a two-item tuple (number of rows and the
number of columns) for the data frame. For a Series, it returns a one-item
tuple. This is the code we use *df.shape* the **first number represents the number of rows and the **second number is the number of columns

*2. Type of Dataset;*
You can use the type ( ) in python to return the type of object.
Checking the type of data frame: type(df)
Checking the type of a column (çhas) within a data frame: df['chas'].dtype

*3. Slicing and Indexing;*
You can use the : operator with the start index on left and end index on
right of it to output the corresponding slice.

(a) Slicing a list: list = [1,2,3,4,5] => list[1:3]
(b) Slicing a Data frame (df) using iloc indexer: df.iloc[:,1:3]

*4. Identifying Unique Elements;*
Using unique ( ) on the column of interest will return a numpy array with unique values of the column.
Extracting all unique values out of ‘’crim” column:
df['crim'].unique()

*5. Value Extraction;*
Using value ( ) on the column of interest will return a numpy array with all
the values of the column.
Extracting values out of ‘’crim” column:
df['PhD'].values

*6. Feature Mean;*
**This is the average** 
Using mean( ) on the data frame will return mean of the data frame across
all the columns.
Extracting the mean:

*7. Feature Median;*
Is the value that separate the upper half of the data to the lower half OR **(Probability 
distribution of the lower half)**
Using median( ) on the data frame will return median values of the data
frame across all the columns.
This is the code for showing the median:
df.median()

*8. Feature Mode;*
Is used to **locate the central tendency of numerical or norminal data**
Using mode( ) on the data frame will return mode values of the data frame

```python
# Merging
import pandas

champion_stats = {'Team':['India','Australia','West Indies',
                          'Pakistan','Sri Lanka'],
                  'ICC_rank':[2,3,7,8,4],
                  'World_champions_Year':[2011,2015,1979,1992,1996],
                  'Points':[874,787,753,673,855]}

match_stats = {'Team':['India','Australia','West Indies',
                       'Pakistan','Sri Lanka'],
               'World_cup_played':[11,10,11,9,8],
               'ODIs_played':[733,988,712,679,662]}

df1=pandas.DataFrame(champion_stats)
df2=pandas.DataFrame(match_stats)

print(df1)
print(df2)
print(pandas.merge(df1,df2,on='Team'))
```

# Different Types of Joins

Joins are used to combine records from two or more tables in a database.

Joins always locate related column values in the two tables.

There are four most commonly used join;

1. Right join ~ Preserves the unmatched rows from the second (right) table, joining them with a NULL in the shape of the first (left) table

2. Left join ~ Returns all rows from the left table, even if there are no matches in the right table

3. Inner join ~ Selects all rows from both participating tables if there is a match between the columns

4. Full outer join ~ Returns all records when there is a match in either left (table1) or right (table2) table records


```python
# Left join
import pandas

world_champions = {'Team':['India','Australia','WestIndies',
                         'Pakistan','Sri Lanka'],
                 'ICC_rank':[2,3,7,8,4],
                 'World_champions_Year':[2011,2015,1979,1992,1996],
                 'Points':[874,787,753,673,855]}

chokers = {'Team':['South Africa','New Zealand',
                   'Zimbabwe'],
           'ICC_rank':[1,5,9],'Points':[895,764,656]}

df1=pandas.DataFrame(world_champions)
df2=pandas.DataFrame(chokers)

print(pandas.merge(df1,df2,on='Team',how='left'))
```

```python
# Right join
import pandas

world_champions = {'Team':['India','Australia','WestIndies',
                         'Pakistan','Sri Lanka'],
                 'ICC_rank':[2,3,7,8,4],
                 'World_champions_Year':[2011,2015,1979,1992,1996],
                 'Points':[874,787,753,673,855]}

chokers = {'Team':['South Africa','New Zealand',
                   'Zimbabwe'],
           'ICC_rank':[1,5,9],'Points':[895,764,656]}

df1=pandas.DataFrame(world_champions)
df2=pandas.DataFrame(chokers)

print(pandas.merge(df1,df2,on='Team',how='right'))
```

```python
# Inner join
import pandas

world_champions = {'Team':['India','Australia','WestIndies',
                         'Pakistan','Sri Lanka'],
                 'ICC_rank':[2,3,7,8,4],
                 'World_champions_Year':[2011,2015,1979,1992,1996],
                 'Points':[874,787,753,673,855]}

chokers = {'Team':['South Africa','New Zealand',
                   'Zimbabwe'],
           'ICC_rank':[1,5,9],'Points':[895,764,656]}

df1=pandas.DataFrame(world_champions)
df2=pandas.DataFrame(chokers)

print(pandas.merge(df1,df2,on='Team',how='inner'))
```

```python
# Fill outer join
import pandas

world_champions = {'Team':['India','Australia','WestIndies',
                         'Pakistan','Sri Lanka'],
                 'ICC_rank':[2,3,7,8,4],
                 'World_champions_Year':[2011,2015,1979,1992,1996],
                 'Points':[874,787,753,673,855]}

chokers = {'Team':['South Africa','New Zealand',
                   'Zimbabwe'],
           'ICC_rank':[1,5,9],'Points':[895,764,656]}

df1=pandas.DataFrame(world_champions)
df2=pandas.DataFrame(chokers)

print(pandas.merge(df1,df2,on='Team',how='outer'))
```
